### 1. 자동미분

```python
import tensorflow as tf
```

```python
w = tf.Variable(2.)
```

```python
def f(w):
    y = w**2
    z = 2*y + 5
    return z
```

```python
## 미분값 출력
with tf.GradientTape() as tape:
    z = f(w)
    
gradients = tape.gradient(z, [w])
print(gradients)
```

### 2. 자동미분을 이용한 선형회귀 구현

```python
### 학습될 가중치 변수 선언
w = tf.Variable(4.0)
b = tf.Variable(1.0)
```

```python
### 가설을 함수로 정의
@tf.function
def hypothesis(x):
    return w*x + b
```

```python
x_test = [3.5, 5, 5.5, 6]
print(hypothesis(x_test).numpy())
```

```python
### 평균 제곱오차를 손실함수로 정의
@tf.function
def mse_loss(y_pred, y):
    return tf.reduce_mean(tf.square(y_pred-y))
```

```python
x = list(range(1, 10))  # 공부한 시간
y = [11, 22, 33, 44, 53, 66, 77, 87, 95] # 공부시간별 성적
```

```python
## 경사하강법(SGD)과 learning_rate 0.01 적용한 optimzer
optimizer = tf.optimizers.SGD(learning_rate=0.01)
```

```python
### 300번에 걸쳐 SGD 수행
for i in range(301):
    with tf.GradientTape() as tape:
        y_pred = hypothesis(x)
        cost = mse_loss(y_pred, y) # 평균제곱오차 계산
        gradients = tape.gradient(cost, [w, b]) # 손실함수에 대한 파라미터 미분값 계산
        
        ## 파라미터 업데이트
        optimizer.apply_gradients(zip(gradients, [w, b]))
        
        if i % 10 == 0:
            print('epoch: {:3} | w의 값 : {:5.4f} | b의 값 : {:5.4} | cost : {:5.6f}'.format(i, w.numpy(), b.numpy(), cost))
            
```

```python
## 예측
x_test = [3.5, 5, 5.5, 6]
print(hypothesis(x_test).numpy())
```

### 3. Keras를 이용한 선형회귀 구현

* keras model 만들기 
    * Sequential로 model 만들고 
    * add를 통해 입력과 출력 벡터의 차원과 같은 정보 추가
    * activation : 어떤 함수를 사용할지 결정, 선형함수일 경우 linear 설정

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 공부하는 시간
y = [11, 22, 33, 44, 53, 66, 77, 87, 95] # 각 공부하는 시간에 맵핑되는 성적
```

```python
model = Sequential()

# 출력 y 차원은 1. 입력 x 차원도 1
# 선형회귀이므로 activation=linear
model.add(Dense(1, input_dim=1, activation='linear'))
```

```python
# 최적화
sgd = optimizers.SGD(learning_rate=0.01)

# 손실함수 만들기
model.compile(optimizer=sgd, loss='mse', metrics=['mse'])

# 학습 : 300번 시도
model.fit(x, y, epochs=300)
```

```python
%matplotlib inline

plt.plot(x, model.predict(x), 'b', x, y, 'k.')  # y는 실제값 dot으로 표시 
plt.show()
```
